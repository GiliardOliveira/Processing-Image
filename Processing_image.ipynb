```
from math import radians
import numpy as np
import matplotlib.pyplot as plt


def main():
    x = np.arange(0, radians(1800), radians(12))
    plt.plot(x, np.cos(x), 'b')
    plt.show()

main()
```

```
%matplotlib inline
import matplotlib.pyplot as plt
from skimage import data
from skimage.color import rgb2gray
original = data.astronaut()
grayscale = rgb2gray(original)

fig, axes = plt.subplots(1,2,figsize=(8,4))
ax= axes.ravel()


ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(grayscale,cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

```
import cv2
img = cv2.imread('ponte.jpg')
for y in range (0,img.shape[0]):
    for x in range(0,img.shape[1]):
      img[y,x] = (255,0,0)
cv2_imshow(img)
```

```
import cv2
img = cv2.imread('ponte.jpg')
for y in range (0,img.shape[0],1):
    for x in range(0,img.shape[1],1):
      img[y,x] = (0,(x*y)%256,0)
cv2_imshow(img)
```

```
import cv2
img = cv2.imread('ponte.jpg')
for y in range (0,img.shape[0]):
    for x in range(0,img.shape[1]):
      img[y,x] = (0,y%256,0)
cv2_imshow(img)
```

```
import cv2
img = cv2.imread('ponte.jpg')
for y in range(0,img.shape[0],10):
  for x in range(0,img.shape[1],10):
    img[y:y+5,x:x+5] = (0,0,255)
cv2_imshow(img)
```

```
import cv2
image = cv2.imread('ponte.jpg')
image[30:50,:] = (255,0,0)
image[100:150,50:100] = (0,0,255)
image[:,200:220] = (0,255,255)
image[150:300,250:350] = (0,255,0)
image[300:400,50:150] = (255,255,0)
image[250:350,300:400] = (255,255,255)
image[70:100,300:450] = (0,0,0)
cv2_imshow(image)
```

```
import numpy as np 
import cv2
imagem = cv2.imread('ponte.jpg')
vermelho = (0,0,255)
verde = (0,255,0)
azul = (255,0,0)

cv2.line(imagem, (0,0),(100,200),verde)
cv2.line(imagem,(300,200),(150,150),vermelho,5)
cv2.rectangle(imagem, (20,20),(120,120),azul)
```
